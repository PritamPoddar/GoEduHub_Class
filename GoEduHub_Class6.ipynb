## Line Graph

```python
# Plotting one line 
Plt.plot([1,2,3],[4,5,1])
Plt.title("Information")
Plt.xlabel("X axis")
Plt.ylabel("Y axis")
Plt.show()
```

```python
# Plotting 2 line 
X = [18,20,26]
Y = [16,21,24]
X1 = [18,21,26]
Y1 = [24,20,16]
Plt.plot(X,Y, "Red", label = "line", linewidth = 4)
Plt.plot(X1,Y1,"Blue", label= "Line", linewidth  =2)
Plt.legend()  # legend will help to show the label
Plt.title("Graph")
Plt.xlabel("X Axis")
Plt.ylabel("Y Axis")
Plt.grid()
Plt.show()
```

## Bar Graph

```python
Plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label="BMW",color='r',width=.1)

Plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label="Audi", color='b',width=.5)

Plt.legend()

Plt.xlabel('Days')
Plt.ylabel('Distance (kms)')
Plt.title('Information')

Plt.show()
```

## Scatter Plot

```python
X = [2,3,4,5,6,7,8,9,10]
Y = [2,3,4,5,6,7,8,9,10]
Plt.scatter(X,Y, color = "Red", label = "line")
Plt.title("Scatter Plot")
Plt.xlabel("X Axis")
Plt.ylabel("Y Axis")
Plt.legend()

Plt.show()
```

```python
X = [2,3,4,5,6,7,8,9,10]
Y = [2,3,5,4,6,7,5,6,8]

X1 = [5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
Y1 = [1,1.5,2,2.5,3,3.5,4,4.5,5,5.5]

Plt.scatter(X,Y, color = "Red", label = "line")
Plt.scatter(X1,Y1, color= "Blue", label = "line")
Plt.title("Scatter Plot")
Plt.xlabel("X Axis")
Plt.ylabel("Y Axis")
Plt.legend()

Plt.show()
```

## Area Plot

```python
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

Plt.plot([],[],color='m', label='Sleeping', linewidth=5)

Plt.plot([],[],color='c', label='Eating', linewidth=5)

Plt.plot([],[],color='r', label='Working', linewidth=5)

Plt.plot([],[],color='k', label='Playing', linewidth=5)

Plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])

Plt.xlabel('x')

Plt.ylabel('y')

Plt.title('Stack Plot')

Plt.legend()

Plt.show()
```

## Pie Chart

```python
slices = [12,22,32,42]

activities = ['sleeping','eating','working','playing']

cols = ['c','m','r','b']

Plt.pie(slices, labels=activities, colors=cols, startangle=180  ) # 180/90/60 or any thing

Plt.title('Pie Plot')

Plt.show()
```

## Working With Multiple Plots

```python
def F(T):
  return NP.exp(-T) * NP.cos(2*NP.pi*T)

T1 = NP.arange(0.0, 5.0, 0.1)

T2 = NP.arange(0.0, 5.0, 0.02)

Plt.subplot(211)  #no of plots, row wise or column wise, 1 2 plot

Plt.plot(T1, F(T1), 'bo', T2, F(T2))

Plt.subplot(212)

Plt.plot(T2, NP.cos(2*NP.pi*T2))

Plt.show()
```

##Iris Datset

```python
import pandas as PD
import numpy as NP
from matplotlib import pyplot as Plt
```

```python
I = PD.read_csv("Iris.csv")
I
```

## Matplot-2D
Scatter Plot

```python
for Species, Iris_Subset in I.groupby("Species"):
  Plt.scatter(Iris_Subset["PetalLengthCm"],Iris_Subset["PetalWidthCm"], label = Species)
Plt.legend()
Plt.show()
```

## Matplot-3D

```python
Fig = Plt.figure()
Ax = Fig.add_subplot(111, projection = "3d")

for Species, Iris_Subset in I.groupby("Species"):
  Ax.scatter(Iris_Subset["PetalLengthCm"],Iris_Subset["PetalWidthCm"], Iris_Subset["SepalLengthCm"], label = Species)
Plt.legend()

Ax.set_xlabel("PetalLength")
Ax.set_ylabel("PetalWidth")
Ax.set_zlabel("SepalLength")

Plt.show()
```
